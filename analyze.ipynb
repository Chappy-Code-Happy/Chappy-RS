```python
cf_problem = pd.read_csv('data/codeforces/problem.csv')
cf_problem.head()
```

```python
## data processing
cf_data = cf_problem[['contestID', 'problemID', 'tags', 'difficulty', 'points', 'solvedCount']]
cf_data.head()
# cc_data.shape
```

```python
## sort by difficulty
cf_data = cf_data.sort_values(by=['difficulty'], ascending=False)
cf_data.head()
```

```python
## count by difficulty
cf_data.value_counts('difficulty')

## analyze difficulty
cf_data['difficulty'].mean() ## 평균 1747.5
cf_data['difficulty'].median() ## 중앙값 1700.0
cf_data['difficulty'].mode() ## 최빈값 800
```

```python
## analyze points
cf_data['points'].mean() ## 평균 1013.5
cf_data['points'].median() ## 중앙값 500.0
cf_data['points'].mode() ## 최빈값 0.0    
```
